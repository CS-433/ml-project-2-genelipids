```python
# Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cells = pd.read_hdf('data/cell_filtered_w500genes.h5', 'df')
```

```python
cells
```

```python
cells_section_12 = cells.loc[(cells['x_ccf'] > 7.4) & (cells['x_ccf'] < 7.8), :]
cells_section_12
```

```python
# genes: note here we have only 500 ST genes, inputation of full transcriptomes still problematic...

plt.figure(figsize=(12, 8))

dot_size = 0.05

sc1 = plt.scatter(cells_section_12['z_ccf'], -cells_section_12['y_ccf'],
            c=cells_section_12['ENSMUST00000109546'], s=dot_size, cmap="Blues", alpha=1)

plt.axis('off')
plt.show()
```

```python
# annotations

plt.figure(figsize=(12, 8))

dot_size = 0.05

plt.scatter(cells_section_12['z_ccf'], -cells_section_12['y_ccf'],
            c=cells_section_12['parcellation_structure_color'], s=dot_size, alpha=1)

plt.axis('off')
plt.show()
```

```python
# celltypes

plt.figure(figsize=(12, 8))

dot_size = 0.05

plt.scatter(cells_section_12['z_ccf'], -cells_section_12['y_ccf'],
            c=cells_section_12['subclass_color'], s=dot_size, alpha=1)

plt.axis('off')
plt.show()
```
